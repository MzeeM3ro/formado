```python
data= pd.read_csv('/content/drive/My Drive/ir04_m_1.csv', sep=",",skiprows=3,encoding='cp932')
df = data.drop(data.index[[0,1,2,3,4]])
df
```

```python
import numpy as np
dataset = df[['新規/短期/国内銀行']]
dataset = dataset.astype(float) #よくわからんけど少数点がながいとchearになるとか
dataset
```

```python
#(0,1)に正規化
dataset -= np.min(np.abs(dataset))
dataset /= np.max(np.abs(dataset))
```

```python
#正規化された乗客数の推移を図示して確認する
plt.plot(dataset)
plt.xlabel("time")
plt.ylabel("y")
plt.title("新規/短期/国内銀行")
plt.show()
```

```python
nmp=dataset.to_numpy()
nmp[1]
```

```python
def create_dataset(dataset, steps_of_history, steps_in_future):
    X, Y = [], []
    for i in range(0, len(dataset)-steps_of_history, steps_in_future):
        X.append(dataset[i:i+steps_of_history])
        Y.append(dataset[i + steps_of_history])
    X = np.reshape(np.array(X), [-1, steps_of_history, 1])
    Y = np.reshape(np.array(Y), [-1, 1])
    return X, Y
```

```python
steps_of_history = 1
steps_in_future = 1

X, Y = create_dataset(nmp, steps_of_history, steps_in_future)
```

```python
from sklearn.model_selection import train_test_split
#分割しましょう
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) #訓練が7割、テストが3割
```

https://qiita.com/Hatomugi/items/bd5a8cadf0c19e3ec06a

```python
#LSTM
#ネットワークの構築
net = tflearn.input_data(shape=[None, steps_of_history, 1])
net = tflearn.lstm(net, n_units=6)
net = tflearn.fully_connected(net, 1, activation='linear')
net = tflearn.regression(net, optimizer='adam', learning_rate=0.001,
        loss='mean_square')

model = tflearn.DNN(net, tensorboard_verbose=0)
model.fit(x_train,y_train, validation_set=0.1, batch_size=1, n_epoch=150)
```

```python
train_predict = model.predict(x_train)
test_predict = model.predict( x_test)

train_predict_plot = np.empty_like(nmp)
train_predict_plot[:, :] = np.nan
train_predict_plot[steps_of_history:len(train_predict)+steps_of_history, :] = \
        train_predict

test_predict_plot = np.empty_like(dataset)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict)+steps_of_history:len(dataset), :] = \
        test_predict

plt.figure(figsize=(8, 8))
plt.title('History={} Future={}'.format(steps_of_history, steps_in_future))
plt.plot(dataset)
plt.plot(train_predict_plot)
plt.plot(test_predict_plot)
```

```python
#GRU
#ネットワークの構築
net = tflearn.input_data(shape=[None, steps_of_history, 1])
net = tflearn.gru(net, n_units=6)
net = tflearn.fully_connected(net, 1, activation='linear')
net = tflearn.regression(net, optimizer='adam', learning_rate=0.001,
        loss='mean_square')

model = tflearn.DNN(net, tensorboard_verbose=0)
model.fit(x_train,y_train, validation_set=0.1, batch_size=1, n_epoch=150)
```

```python
train_predict = model.predict(x_train)
test_predict = model.predict( x_test)

train_predict_plot = np.empty_like(nmp)
train_predict_plot[:, :] = np.nan
train_predict_plot[steps_of_history:len(train_predict)+steps_of_history, :] = \
        train_predict

test_predict_plot = np.empty_like(dataset)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict)+steps_of_history:len(dataset), :] = \
        test_predict

plt.figure(figsize=(8, 8))
plt.title('History={} Future={}'.format(steps_of_history, steps_in_future))
plt.plot(dataset)
plt.plot(train_predict_plot)
plt.plot(test_predict_plot)
```
