```python
from sklearn.datasets import load_breast_cancer #がんの診断結果
from IPython.display import display
from sklearn.model_selection import train_test_split #訓練データとテストデータで過学習防ごうね
from sklearn.preprocessing import StandardScaler
import pandas as pd #おなじみデータフレーム処理君
import seaborn as sns #散布図おなじみさん
from sklearn import linear_model #モデルさんいつもの


data = load_breast_cancer()
```

```python
df_x = pd.DataFrame(
            data = data.data, #列にデータ
            columns=data.feature_names #行に名前
                         )

df_x.describe #内容の確認
```

説明変数が30個あることが分かります

```python
df_y = pd.DataFrame(
            data = data.target, #列にデータ
                         )

df_y.describe #内容の確認
```

```python
#分割しましょう
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3) #訓練が7割、テストが3割
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
 
```

```python
#評価の実行（確率）
df = pd.DataFrame(model.predict_proba(x_test)) 
df = df.rename(columns={0: 'やばい',1: '大丈夫'})
df.describe
```

```python
#評価の実行（判定）
df = pd.DataFrame(model.predict(x_test))
df = df.rename(columns={0: '判定(0:やばい / 1:大丈夫)'})
df.describe
```

```python
#混同行列 窓に書いてます
from sklearn.metrics import confusion_matrix

df = pd.DataFrame(confusion_matrix(y_test,model.predict(x_test).reshape(-1,1), labels=[0,1]))
df = df.rename(columns={0: '予想(やばい)',1: '予想(大丈夫)'}, index={0: '実際(やばい)',1: '実際(だいじょうぶ)'})
df
 
```

```python
#評価の実行（正答率）
model.score(x_test,y_test)
```
